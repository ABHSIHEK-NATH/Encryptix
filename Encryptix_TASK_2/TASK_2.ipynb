# As we are going to predict movie ratings based on Name, Year, Duration, Votes, Rating. So we need to remove null values from the dataset so that the result doesn't get influenced.


```python
movie_df.dropna(subset=['Name','Year','Duration','Votes','Rating'],inplace=True)
movie_df.isna().sum()
```

# * Removing parentheses from 'Year' column and convert to integer
# * Removing commas from 'Votes' column and convert to integer
# * Removing min from 'Duration' column and convert to integer

```python
movie_df['Year'] = movie_df['Year'].str.strip('()').astype(int)
```

```python
movie_df['Votes'] = movie_df['Votes'].str.replace(',', '').astype(int)
```

```python
movie_df['Duration'] = movie_df['Duration'].str.replace('min', '').astype(int)
```

# Future Engineering

```python
movie_df.drop(['Name','Director','Actor 1','Actor 2','Actor 3'], axis=1,inplace=True)
movie_df.head()
```

```python
X = movie_df[['Year','Duration','Votes']]
y = movie_df['Rating']
```

# Splitting the data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1000)
```

# Building the Model

-> Creating a pipeline with standard scaling and SGD regression

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('sgd', SGDRegressor(max_iter=10000, random_state=1000))
])
```

```python
pipeline.fit(X_train, y_train)
```

# Predicting Rating on the test set

```python
y_pred_pipeline = pipeline.predict(X_test)
```

# Model Evaluation

-> Evaluatinf Metrics for the Pipeline

```python
mae_pipeline = mean_absolute_error(y_test, y_pred_pipeline)
mse_pipeline = mean_squared_error(y_test, y_pred_pipeline)
r2_pipeline = r2_score(y_test, y_pred_pipeline)
```

```python
print("Pipeline Mean Absolute Error:", mae_pipeline)
print("Pipeline Mean Squared Error:", mse_pipeline)
print("Pipeline R-squared:", r2_pipeline)
```

# Deploying the Model

```python
new_input = pd.DataFrame({'Year': [2023], 'Duration': [120], 'Votes': [10000]})
predicted_rating = pipeline.predict(new_input)

print("Predicted Rating:", predicted_rating)
```
